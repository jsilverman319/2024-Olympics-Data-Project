```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_olympics = pd.read_csv('/Users/jahansilverman/Documents/CSVs for Data Science Projects/olympics.csv', header=0)
```

```python
df_olympics.head()
```

```python
df_olympics.dtypes
```

```python
df_olympics.rename(columns={'gdp': 'gdp_per_capita', 'population': 'population_millions'}, inplace=True)
```

```python
df_olympics.head()
```

```python
sns.scatterplot(data=df_olympics, x='population_millions', y='total', color='Orange')

plt.title('Total Medals vs. Population (Millions)')
plt.xlabel('Population (Millions)')
plt.ylabel('Total Medals')

plt.show()
```

```python
sns.scatterplot(data=df_olympics, x='gdp_per_capita', y='total', color='Blue')

plt.title('Total Medals vs. GDP Per Capita')
plt.xlabel('GDP Per Capita')
plt.ylabel('Total Medals')

plt.show()
```

```python
df_olympics_bar = df_olympics.groupby('region').agg({'total':sum}).reset_index().sort_values(by='total',
                                                                                             ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation=20)
sns.barplot(x='region',y='total',data=df_olympics_bar).set(xlabel=None, ylabel=None)
plt.title('Total Medals vs. Region')

plt.show()
```

```python
df_olympics_bar = df_olympics.groupby('country').agg({'total':sum}).reset_index().sort_values(by='total',
                                                                                              ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation=20)
sns.barplot(x='country',y='total',data=df_olympics_bar.head(20)).set(xlabel=None, ylabel=None)
plt.title('Total Medals vs. Country')

plt.show()
```

```python
df_olympics_bar = df_olympics.groupby('region').agg({'gold':sum}).reset_index().sort_values(by='gold',
                                                                                            ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation=20)
sns.barplot(x='region',y='gold',data=df_olympics_bar).set(xlabel=None, ylabel=None)
plt.title('Gold Medals vs. Region')

plt.show()
```

```python
df_olympics_bar = df_olympics.groupby('country').agg({'gold':sum}).reset_index().sort_values(by='gold',
                                                                                             ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation=20)
sns.barplot(x='country',y='gold',data=df_olympics_bar.head(20)).set(xlabel=None, ylabel=None)
plt.title('Gold Medals vs. Country')

plt.show()
```
